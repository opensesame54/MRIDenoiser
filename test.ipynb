```python
from google.colab import files
import io
import os
import shutil
import zipfile

uploaded = files.upload()
zip_name = next((name for name in uploaded.keys() if name.endswith('.zip')), None)
if zip_name is None:
    raise RuntimeError('Please upload pcap_proj.zip')

if os.path.exists('/content/pcap_proj'):
    shutil.rmtree('/content/pcap_proj')

with zipfile.ZipFile(io.BytesIO(uploaded[zip_name]), 'r') as zf:
    zf.extractall('/content')

if not os.path.exists('/content/pcap_proj/CMakeLists.txt'):
    raise RuntimeError('Uploaded zip does not contain pcap_proj/CMakeLists.txt')

print('Project extracted to /content/pcap_proj')
```
